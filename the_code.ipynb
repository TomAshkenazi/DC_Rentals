```python
list_cards = soup.find_all('div', class_='list-card-info')
for listing in list_cards:
    try:
        table = []
        address = listing.find('address', class_="list-card-addr").text
        price_monthly = listing.find('div', class_="list-card-price")
        if price_monthly is not None:
            monthly_price = price_monthly.text
        else:
            monthly_price = None
        search = listing.find('ul', class_="list-card-details")
        lis = search.find_all('li')
        table.append(lis)
        also_price = lis[0].text
        if price_monthly is not None:
            sqft = lis[2].text
        else:
            sqft = None
        house_dict = {"address": address, "price_monthly": monthly_price, "also_price": also_price, "sqft": sqft}
        apartments_20009.append(house_dict)
                
        if (address):
            print('-------------')
            print(address)
            print(price_monthly)
            print(also_price)
            print(sqft)
    
    except IndexError as e:
        print(e)
```
